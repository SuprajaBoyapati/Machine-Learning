# Importing Libraries

```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

# Reading the winemag dataset

```python
df = pd.read_csv("/content/winemag-data.csv", index_col=0)
```

# Printing the first five elements form dataset

```python
df.head()
```

# Using Counter on Variety column.

```python
C=Counter(df['variety'])
```

# Printing the elements from the counter

```python
C
```

```python
len(C) # we have total 41 types of wines. We have to plot 41 word clods.
```

# Using for loop to print the top five most used for each wine.

```python
for i in C:
  
  new_df=df[df['variety']==i]
  new_text=''
  for j in new_df.description:
    new_text+=j
  print(" The top five most words used in the wine",i ,"are")
  wordcloud = WordCloud(background_color="white",max_words=5).generate(new_text)
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```
